```python
def local_heating(heat_type, heat_val, plev, lon_min, lon_max, lat_min, lat_max, lon_cen):
    heating = xr.open_dataset("/scratch/w40/gp8431/run/"+heat_type+"_amp_"+heat_val+"/plevel_avg_"+heat_type+".nc", decode_times = False).local_heating
    heat_slice = heating.sel(lat=slice(lat_min, lat_max), lon=slice(lon_min, lon_max), pfull=plev)
    heat_peak = heating.sel(lat=0.0, lon=lon_cen, pfull=plev, method="nearest")
    return heating, heat_slice, heat_peak
```

```python
gc, gc_slice, gc_peak = local_heating('gh', '-10.0',300.0, 85., 105., -10., 10., 95.)
```
